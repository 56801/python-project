```python
ax = sns.countplot(x= "Churn",data = df)
ax.bar_label(ax.containers[0])
plt.title("count of customers by churn")
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"], labels = gb.index, autopct = "%1.2f%%")
plt.title("percentage of churn customers")
plt.show()
```

```python


# Count values based on 'Gender' and 'Churn'
gender_churn_counts = df.groupby(['gender', 'Churn']).size().unstack()

# Print the counts
print(gender_churn_counts)

# Plotting the grouped bar plot
gender_churn_counts.plot(kind='bar', figsize=(7, 5), color=['#1f77b4', '#ff7f0e'])  # Different colors for each bar
plt.title('churn by gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep the labels horizontal
plt.legend(title='Churn')  # Add legend for Churn categories
plt.show()
```

```python


# Count values in the 'SeniorCitizen' column
senior_citizen_counts = df['SeniorCitizen'].value_counts()

# Plotting the bar plot
ax = senior_citizen_counts.plot(kind='bar', color=['#1f77b4', '#ff7f0e'], figsize=(7, 5))  # Different colors for each bar

# Adding count labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', fontsize=9, color='black', xytext=(0, 5), 
                textcoords='offset points')

# Plot title and labels
plt.title('Count of Senior Citizens')
plt.xlabel('Senior Citizen')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Keep the labels horizontal
plt.show()
```

```python


# Count values based on 'SeniorCitizen' and 'Churn'
senior_churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Calculate the percentage
senior_churn_percentage = senior_churn_counts.div(senior_churn_counts.sum(axis=1), axis=0) * 100

# Plotting the stacked bar chart
ax = senior_churn_percentage.plot(kind='bar', stacked=True, figsize=(7, 5), color=['#1f77b4', '#ff7f0e'])  # Different colors for 'Churn'

# Adding percentage labels to each bar
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f%%', label_type='center')  # Label format to show percentage

# Plot title and labels
plt.title('SeniorCitizen vs Churn Distribution (Percentage)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)  # Keep the labels horizontal
plt.legend(title='Churn')  # Add legend for Churn categories
plt.show()
```

```python


# List of columns to create countplots for
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the figure and axes for subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 16))  # 3x3 grid of subplots

# Loop through each column and create a countplot with hue='Churn'
for i, col in enumerate(columns):
    row = i // 3
    col_pos = i % 3
    sns.countplot(x=col, data=df, ax=axes[row, col_pos], palette='dark:skyblue', hue=df['Churn'])

    # Set the title and annotate the counts
    axes[row, col_pos].set_title(f'Countplot of {col}')
    for p in axes[row, col_pos].patches:
        axes[row, col_pos].annotate(f'{p.get_height()}', 
                                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                                     ha='center', va='baseline', fontsize=10, color='black', 
                                     xytext=(0, 5), textcoords='offset points')

# Adjust layout
plt.tight_layout()
plt.show()
```

The subplots visualize customer churn across various services. Customers who use phone services, online security, and tech support have lower churn rates, with a significant portion retaining services. On the other hand, for customers without internet service or multiple lines, the churn rate is notably higher. Fiber optic users also seem to have a higher tendency to churn compared to DSL users. These insights suggest that customers with more comprehensive services tend to remain loyal, while others may cancel more frequently.
